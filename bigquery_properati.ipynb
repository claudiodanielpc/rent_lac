```python
from google.cloud import bigquery
from google.colab import auth
# Autenticate
auth.authenticate_user()
```

```python
# Create the client
client = bigquery.Client(project='clasificador-pgd')
```

### BigQuery example using Github repo

```python
#Query using github repo
query = """
SELECT
  *
FROM
  `bigquery-public-data.github_repos.sample_contents`
LIMIT 5;
"""

# dataframe
df = client.query(query).to_dataframe()

df.head(5)
```
