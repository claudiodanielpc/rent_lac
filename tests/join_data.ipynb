```python
import pandas as pd
import os
```

```python
path="tests/scraping_output/raw"
```

```python
# Filtrar archivos que empiezan con "bronze" y son csv
files = [
    os.path.join(path, f)
    for f in os.listdir(path)
    if f.startswith("bronze") and f.endswith(".csv")
]

# Leer y concatenar
df = pd.concat(
    [pd.read_csv(f) for f in files],
    ignore_index=True
)

print(f"Archivos concatenados: {len(files)}")
```

```python
df
```

```python
#Salvar csv como total_scraping

df.to_csv(path + "/total_scraping.csv",  index=False, encoding="utf-8-sig")
```
