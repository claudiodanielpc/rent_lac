```python
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
path="scraping_output_round2/raw"
```

```python
# Filtrar archivos que empiezan con "bronze" y son csv
files = [
    os.path.join(path, f)
    for f in os.listdir(path)
    if f.startswith("bronze") and f.endswith(".csv")
]

# Leer y concatenar
df = pd.concat(
    [pd.read_csv(f) for f in files],
    ignore_index=True
)

print(f"Archivos concatenados: {len(files)}")
```

```python
df
```

```python
#Salvar csv como total_scraping

df.to_csv(path + "/total_scraping_round2.csv",  index=False, encoding="utf-8-sig")
```

```python
###Pegar las rondas

round1 = pd.read_csv("scraping_output/raw/total_scraping.csv",low_memory=False)
round2 = pd.read_csv("scraping_output_round2/raw/total_scraping_round2.csv",low_memory=False)
total = pd.concat([round1, round2], ignore_index=True)
total.to_csv("scraping_output/raw/total_scraping_rounds.csv", index=False, encoding="utf-8-sig")
```
